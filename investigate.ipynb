```python
import pickle as pkl
import matplotlib.pyplot as plt
```

```python
act_32 = pkl.load(open('nsamples=32.pkl', 'rb'))
# act_128 = pkl.load(open('nsamples=128.pkl', 'rb'))
act_512 = pkl.load(open('nsamples=512.pkl', 'rb'))

act_og_32 = pkl.load(open('og_llama_nsamples=32.pkl', 'rb'))
act_og_512 = pkl.load(open('og_llama_nsamples=512.pkl', 'rb'))
```

```python
for idx, (k, v) in enumerate(act_32.items()):
    if (idx % 4) == 0:
        out_32 =  (v['out'][1] / v['out'][0]).cpu().numpy()
        og_32 =  (act_og_32[k]['out'][1] / act_og_32[k]['out'][0]).cpu().numpy()

    #     out_128 =  (act_128[k]['out'][1] / act_128[k]['out'][0]).cpu().numpy()
        out_512 =  (act_512[k]['out'][1] / act_512[k]['out'][0]).cpu().numpy()
        og_512 =  (act_og_512[k]['out'][1] / act_og_512[k]['out'][0]).cpu().numpy()

        fig, ax = plt.subplots(figsize =(6, 3))
        ax.hist(out_32, bins=100,  label='32', histtype='step', color='tab:red')
        ax.hist(og_32, bins=100, label='OG-32', linestyle=('dashed'), histtype='step', color='tab:cyan')

    #     ax.hist(out_128, bins=100, label='128', histtype='step', color='tab:cyan')
        ax.hist(out_512, bins=100, label='512', histtype='step', color='tab:orange')
        ax.hist(og_512, bins=100, label='OG-512', linestyle=('dashed'), histtype='step', color='tab:grey')
        
        ax.set_title(k)
        ax.set_xscale('log')
        plt.legend()
        plt.show()
```
